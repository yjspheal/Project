```python
import re
from datetime import datetime,timedelta
import time
import matplotlib as mpl
import matplotlib.pylab as plt
import pandas as pd

titlelist=['Aroha','Any Song','GGang','Blueming','Left&Right']
singerlist=['조정석','지코','비','아이유','세븐틴']
datelist=['2020.03.27','2020.01.13','2017.12.01','2019.11.18','2020.06.22']
genrelist=['발라드','힙합','힙합','메탈','댄스']
lyricslist=['어두운 불빛아래\n촛불 하나','아무 노래나 일단 틀어\n아무거나 신나는 걸로','화려한 조명이\n나를 감싸네','엄지손가락으로\n장미꽃을 피워','결승선을 넘을 때 세리머니']
timelist=[1,4,5,2,4]
pl=pd.DataFrame(data=[['','','','','']],columns=['제목','가수','날짜','장르','횟수'])
df=''

class NNameError(Exception):
    def __init__(self,msg):
        self.message=msg
        
class admin:
    def __init__(self,title,singer,date,genre,lyrics):
        self.title=title
        self.singer=singer
        self.date=date
        self.genre=genre
        self.lyrics=lyrics
        
    def add_song(self):
        global titlelist; global singerlist; global datelist; global genrelist; global timelist; global lyricslist
        global df
        
        titlelist.append(self.title)
        singerlist.append(self.singer)
        datelist.append(self.date)
        genrelist.append(self.genre)
        lyricslist.append(self.lyrics)
        timelist.append(0)
        
        songset=list(zip(titlelist,singerlist,datelist,genrelist,timelist))
        df=pd.DataFrame(data=songset,columns=['제목','가수','날짜','장르','횟수'])
        
    def songlist():
        global df 
        print(df)
    
        
class user:
    def __init__(self,name):
        self.name=name
        
    def home(self):
        sel=input('1.차트 보기 2.검색 3.플레이리스트')
        if sel=='1':
            user.chart(self)
        elif sel=='2':
            user.search(self)
        elif sel=='3':
            user.playlist(self)
        else:
            print('1,2,3 중에 하나를 입력해주세요')
            user.home(self)
            
    def chart(self):
        global titlelist; global singerlist; global datelist; global genrelist
        global df
        
        selchart=input('1.인기 차트 2.최신곡')
        if selchart=='1':
            p_chart=df.sort_values(by='횟수',ascending=False)
            p_chart=p_chart.reset_index(drop=True)
            print(p_chart)
            tx=p_chart['제목'].values.tolist()
            ty=p_chart['횟수'].values.tolist()
            plt.plot(tx,ty)
            plt.xlabel('Title')
            plt.ylabel('Played Time')
            plt.show()
            res=input('1.플레이리스트에 담기 2.담지 않기')
            if res=='1':
                user.add_playlist(p_chart)
            elif res=='2':
                pass
            print()
        elif selchart=='2':
            i=0
            dl=[]; tl=[]; sl=[]; gl=[]
            for date in datelist:
                a=datetime.strptime(date,'%Y.%m.%d')
                if a>datetime.now()-timedelta(days=14):
                    dl.append(date)
                    tl.append(titlelist[i])
                    sl.append(singerlist[i])
                    gl.append(genrelist[i])
                i+=1
            nset=list(zip(tl,sl,dl,gl))
            n_chart=pd.DataFrame(data=nset,columns=['제목','가수','날짜','장르'])
            n_chart=n_chart.reset_index(drop=True)
            print(n_chart)
            res=input('1.플레이리스트에 담기 2.담지 않기')
            if res=='1':
                user.add_playlist(n_chart)
            elif res=='2':
                pass
            print()
            
        print('\n홈으로 돌아갑니다.')
        user.home(self)
    
    def search(self):
        global titlelist; global singerlist; global datelist; global genrelist; global timelist
        songset=list(zip(titlelist,singerlist,datelist,genrelist,timelist))
        A=pd.DataFrame(data=songset,columns=['제목','가수','날짜','장르','횟수'])
        print('곡을 검색합니다\n')
        while 1:
            ct=input('검색 기준 : 1.제목 2.가수 3.장르')
            if ct=='1':
                Tn=input('제목')
                B=A[A.제목==Tn]
                print(B)
                res=input('1.플레이리스트에 담기 2.담지 않기')
                if res=='1':
                    user.add_playlist(B)
                elif res=='2':
                    pass
                print()
            elif ct=='2':
                Sn=input('가수')
                B=A[A.가수==Sn]
                print(B)
                res=input('1.플레이리스트에 담기 2.담지 않기')
                if res=='1':
                    user.add_playlist(B)
                elif res=='2':
                    pass
                print()
            elif ct=='3':
                Gn=input('장르')
                B=A[A.장르==Gn]
                print(B)
                res=input('1.플레이리스트에 담기 2.담지 않기')
                if res=='1':
                    user.add_playlist(B)
                elif res=='2':
                    pass
                print()
            else:
                print('홈으로 돌아갑니다.')
                user.home(self)
                break
                
    def add_playlist(D):
        global pl
        pl=pd.concat([pl,D])
        pl=pl.reset_index(drop=True)
        print('플레이리스트에 담겼습니다.')
                
    def playlist(self):
        global titlelist; global singerlist; global datelist; global genrelist; global timelist
        global pl; global df
        songset=list(zip(titlelist,singerlist,datelist,genrelist,timelist))
        df=pd.DataFrame(data=songset,columns=['제목','가수','날짜','장르','횟수'])
        pl=pl.drop(0);
        pl.drop('횟수', axis=1, inplace=True)
        
        while 1:
            print('현재 플레이리스트 :\n',pl)
            psel=input('플레이리스트를 1.수정합니다. 2.재생합니다.')
            if psel=='1':
                res=input('1.곡 삭제 2.곡 위치 바꾸기')
                if res=='1':
                    rs=input('몇 번 곡을 삭제하시겠습니까?')
                    pl=pl.drop(int(rs))
                    
                elif res=='2':
                    n1,n2=input('순서를 바꿀 곡(n1 n2)').split(' ')
                    pl.loc[100]=[0,0,0,0]
                    pl.loc[100]=pl.loc[int(n1)]
                    pl.loc[int(n1)]=pl.loc[int(n2)]
                    pl.loc[int(n2)]=pl.loc[100]
                    pl=pl.drop(100)
                
            elif psel=='2':
                pl2=pl.set_index('제목')
                for i in range(len(pl.index)):
                    plz=pl.index[i]
                    plz2=pl2.index[i]
                    print(pl.loc[plz]['제목'],'-',pl.loc[plz]['가수'])
                    for j in range(len(titlelist)):
                        if plz2==titlelist[j]:
                            print(lyricslist[j])
                            timelist[j]+=1
                            
                    for k in range(20):
                        print('-',end='')
                        time.sleep(0.1)
                    print()
                songset=list(zip(titlelist,singerlist,datelist,genrelist,timelist))
                df=pd.DataFrame(data=songset,columns=['제목','가수','날짜','장르','횟수'])
                print('홈으로 돌아갑니다.')
                user.home(self)
                break
            else:
                print('홈으로 돌아갑니다.')
                user.home(self)
                break
                
def input_song():
    while 1:
        try:
            title, singer, date, genre,lyrics=input('(영어)제목/가수/날짜/장르/가사 ').split('/')
            s=admin(title,singer,date,genre,lyrics)
            s.add_song()
        except ValueError:
            sel=input('1.곡 목록 보기 2.로그인 화면으로')
            if sel=='1':
                admin.songlist()
            else:
                print('로그인 화면으로 돌아갑니다.')
                break
        
    
    
def input_name():
    while 1:
        nname=input('별명을 입력해주세요(특수문자 불가): ')
        if nname=='':
            raise NNameError('한 글자 이상의 별명을 입력해주세요(enter)')
            break
        elif re.match(r'\w+',nname)==None:
            print('올바른 닉네임을 입력해주세요')    
        else:
            return nname
            break

            
if __name__=="__main__":
    while 1:
        try:
            name=input_name()
        except NNameError as n:               
            print(n.args[0],end='')
            k=input()
            if k=='$^^$$^^$':
                print('관리자 권한으로 로그인합니다.')
                input_song()
        else:
            print('%s님이 로그인 하였습니다.'%name)
            print()
            user.home(name)
            break
            
```
