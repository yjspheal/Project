```python
import numpy as np

phar = pd.read_csv("서울시 약국 인허가 정보.csv", encoding = 'euc-kr')
phar = phar[phar['영업상태명'] != '폐업']

f_phar = pd.read_csv("외국어 가능 약국.csv", encoding = 'euc-kr')
f_phar
```

```python
area = ['중구','종로구', '은평구', '중랑구', '강남구', '서초구', '송파구', '관악구', '금천구', '영등포구', '강서구', '강북구', '노원구','강동구', '양천구', '성동구', '동대문구', '서대문구', '동작구', '구로구', '용산구', '성북구', '도봉구', '마포구']
count_phar = []
count_fphar = []

for a in area:
    count = 0
    for spec in phar['도로명주소']:
        if type(spec) == float:
            continue
        elif spec.split()[1] == a:
            count += 1
            
    count_phar.append(count)
    
for a in area:
    count = 0
    for spec in f_phar['Unnamed: 1']:
        if spec == a:
            count += 1
            
    count_fphar.append(count)
    
print(count_phar)
print(count_fphar)
print(sum(count_fphar)/sum(count_phar) * 100)  # 외국어 가능 약국 전체 비율

```

```python
plt.figure(figsize =(14,8))
plt.bar(area, count_phar, color = "orange")
plt.plot(area, count_fphar, c = "b")
```
